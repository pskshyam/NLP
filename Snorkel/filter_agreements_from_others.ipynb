```python
def preprocess(text):        
    text = text.replace('\n',' ')
    text = re.sub(r'[^.0-9a-zA-Z]', ' ', text)
    text = re.sub(r'[^.\w\s]', ' ', text)
    text = text.lower()
    return text
```
